```python
# Import dependencies
import pandas as pd
import numpy as np
from numpy import array
import matplotlib.pyplot as plt
import warnings
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

%matplotlib inline
```

```python
data_LinReg = pd.read_csv('total_profit_LinReg.csv')
```

```python
data_LinReg.head()
```

```python
Lin_Reg_df = pd.DataFrame(data_LinReg)
Lin_Reg_df = Lin_Reg_df.iloc[65:]
Lin_Reg_df.head()
```

```python
#Create new Dataframe, set 'Date' as index, drop column named 'Date' 
time_indexed_Lin_Reg_df = Lin_Reg_df.set_index(Lin_Reg_df['Date'])
time_indexed_Lin_Reg_df = time_indexed_Lin_Reg_df.drop(['Date'], axis=1)
time_indexed_Lin_Reg_df = time_indexed_Lin_Reg_df.rename(columns={'cum_profit':'LinReg'})
time_indexed_Lin_Reg_df.head(10)
```

```python
data_LSTM = pd.read_csv('total_profit_LSTM.csv')
data_LSTM.head()
```

```python
LSTM_df = pd.DataFrame(data_LSTM)
# LSTM_df = LSTM_df.iloc[1:]
LSTM_df.head()
```

```python
#Create new Dataframe, set 'Date' as index, drop column named 'Date' 
time_indexed_LSTM_df = LSTM_df.set_index(LSTM_df['Date'])
time_indexed_LSTM_df = time_indexed_LSTM_df.drop(['Date'], axis=1)
time_indexed_LSTM_df = time_indexed_LSTM_df.rename(columns={'Profit':'LSTM'})
time_indexed_LSTM_df.head(10)
```

```python
fig, ax = plt.subplots()
fig.suptitle("Total Profit", fontsize=16, fontweight="bold")
ax.set_xlabel("Date")
ax.set_ylabel("Total Profit ($)")
ax.plot(time_indexed_LSTM_df.index, time_indexed_LSTM_df["LSTM"], '-b')
ax.plot(time_indexed_Lin_Reg_df.index, time_indexed_Lin_Reg_df["LinReg"], '-r')

ax.set_xticks(['2018-01-01', '2019-01-01'])

plt.legend(loc='lower right')


plt.show()
```

```python
data_SPX = pd.read_csv('spx_futures.csv')
SPX_df = pd.DataFrame(data_SPX)
SPX_df = SPX_df.iloc[2052:]
SPX_df.head()
```

```python
#Create new Dataframe, set 'Date' as index, drop column named 'Date' 
time_indexed_SPX_df = SPX_df.set_index(SPX_df['Date'])
time_indexed_SPX_df = time_indexed_SPX_df.drop(['Date'], axis=1)
# time_indexed_SPX_df = time_indexed_SPX_df.rename(columns={'Profit':'LSTM'})
time_indexed_SPX_df.head(10)
```

```python
fig, ax = plt.subplots()
fig.suptitle("SPX Stock Price", fontsize=16, fontweight="bold")
ax.set_xlabel("Date")
ax.set_ylabel("SPX Price ($)")
ax.plot(time_indexed_SPX_df.index, time_indexed_SPX_df["SPX"], '-g')


ax.set_xticks(['2018-01-01', '2019-01-01'])

# plt.legend(loc='lower right')


plt.show()
```
